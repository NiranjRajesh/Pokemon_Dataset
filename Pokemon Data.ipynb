```python
import requests
from bs4 import BeautifulSoup


r=requests.get("https://pokemondb.net/pokedex/national")
c=r.content

soup=BeautifulSoup(c,"html.parser")

all=soup.find_all("div",{"class":"infocard"})
poke_dex=[]

for item in all:
    dict={}
    dict["Pokemon"]= item.find_all("a",{"class":"ent-name"})[0].text
    types=item.find_all("small")[1].find_all("a")
    init_itr=True
    for newitem in types:
        if init_itr:
            typing=newitem.text
        elif not init_itr:
            typing=typing+'/'+newitem.text
        init_itr=False  
        dict['Types']=typing
        
    poke_dex.append(dict)        
        

poke_dex

        
        
        


    
    
```

```python
import pandas as pd

df=pd.DataFrame(poke_dex)
df
```
